## Popular Names for Girls from 2011-2021

In order to find the most popular (max) name for a girl in 2011, I first filtered the dataframe to include 'FEMALE' and '2011'. The dataset df includes different ethnicities so there are duplicates of the same name. For example, the name EMMA is found under WHITE-NONHISPANIC and under ASIAN & PACIFIC ISLANDER. So it is important to group the dataset by name and sum the count. Once I have the number of count for each unique name, I can then find the maximum value and output the name with the max value along with the count. I will repeat this process for each year frm 2011 to 2021

#### 2011

```python
# Filter the data frame df to only include rows of 'FEMALE' under the column 'Gender',
# and include rows of '2011' under the column 'Year of Birth'
girls_2011 = df[(df['Gender'] == 'FEMALE') & (df['Year of Birth'] == 2011)]
print(girls_2011.head())

# Group by name and sum the counts. Creates a new index for the count of names from 'girls_2011' df
girls_2011_grouped = girls_2011.groupby('Child\'s First Name')['Count'].sum().reset_index()
print(girls_2011_grouped.head())

# Find the name with the maximum count. Select the row and position that contains the max number
max_count_row_11 = girls_2011_grouped.loc[girls_2011_grouped['Count'].idxmax()]

# Extract the name and the count
most_popular_name_2011 = max_count_row_11['Child\'s First Name']
max_count_2011 = max_count_row_11['Count']


print(f"Most popular female name in 2011: {most_popular_name_2011}")
print(f"Count: {max_count_2011}")
```

## Graph of Most Popular Girl Names from 2011 to 2021

I created a list containing the years and another list that contains the max count from each year. I then created a bar plot of the max count from the years 2011-2016. I manually assigned colors for each year that is associated with the most popular name for that particular year. This is useful for creating distinctions between the bars in the bar plot because some of the years contain the same name (Sophia was a popular name in 2012, 2013, and 2014), so I assigned the color 'teal' for those years.

In this case, I needed to customize the legend to follow the bar graph. Without the legend, it would be confusing to understand the graph because the legend includes the names and the assigned colors to match the bar plot.

```python
# List of years from 2011-2021
years = ["2011", "2012", "2013", "2014", "2015", "2016",
         "2017", "2018", "2019", "2020", "2021"]
print(years)

# Assign colors for the most popular name for each year
# ISABELLA: steelblbue
# SOPHIA: teal
# OLIVIA: mediumaquamarine
# EMMA: cornflowerblue
colors = ['steelblue','teal','teal','teal','mediumaquamarine','mediumaquamarine',
          'cornflowerblue', 'cornflowerblue', 'cornflowerblue', 'cornflowerblue',
          'cornflowerblue']

# List of max count of names for each year
count = [4744, 4374, 4406, 4254, 589, 559, 561, 494, 490, 433, 426]
print(count)

# Reverse the order of the lists for descending years on the plot
years.reverse()
count.reverse()
colors.reverse()

plt.style.use('fivethirtyeight')
# Create bar plot
plt.barh(years, count, color=colors)

# Labels
plt.xlabel('Count of Names')
plt.ylabel('Year')
plt.title('Most Popular Girl Names 2011-2021')

# Create custom legend handles for the names
import matplotlib.patches as mpatches
olivia_patch = mpatches.Patch(facecolor='mediumaquamarine', label='Olivia')
sophia_patch = mpatches.Patch(facecolor='teal', label='Sophia')
isabella_patch = mpatches.Patch(facecolor='steelblue', label='Isabella')
emma_patch = mpatches.Patch(facecolor='cornflowerblue', label='Emma')

# Legend handle that stores the labels
legend_handles = [olivia_patch, sophia_patch, isabella_patch, emma_patch]

plt.tight_layout()
# pass the custom legend handles to handles
plt.legend(handles = legend_handles)
plt.show()
```

#### Summarize Findings for Most Popular Girl Names:

Popular girl names varied throughout the years while some names were consistent. For instance, Sophia was a popular girl name from 2012 up to 2014. Olivia, was most popular in 2015 and 2016 while Emma lasted the longest for most popular girl name from 2017 to 2021. Looking at the graph I noticed that the count for popular girl names significantly dropped at 2015 and continued to decrease throughout time.

## Top 10 Names for Boys in 2020

In this context, I want to understand the relationship between gender and the year. I will be exploring the top 10 popular names for boys in the year 2020. I will filter the dataframe by 'MALE' and the year '2020'.

```python
# Top names for boys in 2020
boys_2020 = df[(df['Gender'] == 'MALE') & (df['Year of Birth'] == 2020)].head(10)
print("\nTop 10 boy names in 2020:")
print(boys_2020[['Child\'s First Name', 'Count', 'Rank']])
```

```python
# Plot: Bar chart for Top 10 Names for Boys in 2020
fig, ax = plt.subplots(figsize=(10, 6))

# Create a colormap
cmap = plt.cm.viridis

# Manually scale the count data to the range [0, 1]
min_count = boys_2020['Count'].min()
max_count = boys_2020['Count'].max()
scaled_counts = (boys_2020['Count'] - min_count) / (max_count - min_count)

# Get colors for each bar by applying the colormap to the scaled counts
colors = cmap(scaled_counts)

ax.bar(boys_2020['Child\'s First Name'], boys_2020['Count'], color=colors)
# Labels
ax.set_title('Top 10 Boy Names in 2020')
ax.set_xlabel('Name')
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Summary Findings:

The name Liam has the highest count for 2020. This suggests over 350 boys were named Liam.

## Trend of The Name Kayla Over Time

In this data analysis, I want to find the popularity of the name Kayla over time from 2011 to 2021. I will first filter the dataframe to include 'KAYLA' and then get the total count of the name 'KAYLA' for each year.

```python
kayla_data = df[(df['Child\'s First Name'] == 'KAYLA')]
yearly_kayla = kayla_data.groupby('Year of Birth')['Count'].sum()
print(yearly_kayla)
```

```python
# Plot: Line chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_kayla, marker='o', color='violet')

#Label
plt.title("Kaylas' Popularity Over Time")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

## Trend of The Name Grace Over time

```python
grace_data = df[(df['Child\'s First Name'] == 'GRACE') & (df['Gender'] == 'FEMALE')]
yearly_grace = grace_data.groupby('Year of Birth')['Count'].sum()
print(yearly_grace)
```

```python
# Plot: Line chart
plt.style.use("ggplot")
plt.figure(figsize=(10, 6))
plt.plot(yearly_grace, marker='o', color='salmon')

# Labels
plt.title("Grace's Popularity Over Time")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

## Combine Grace and Kayla's Plot

```python
# Kayla's plot
plt.plot(yearly_kayla, marker='o', color='violet')

# Grace's plot
plt.plot(yearly_grace, marker='o', color='salmon')
# Labels
plt.title("Grace's and Kayla's Popularity Over Time")
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(['Kayla', 'Grace'])
plt.show()
```
