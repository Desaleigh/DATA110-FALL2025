##**Task 1: Basic Bar Chart for Crime Rates on 2023**

```python
crime_type =["Aggravated Assault","Burglary","Larceny", "Motor Vehicle Theft", "Murder", "Rape", "Robbery"]
offense_rate =[264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
plt.barh(y = crime_type, width = offense_rate)
plt.xlabel("Offense Rate")
plt.ylabel("Crime Type")
plt.title("Crime Rates in 2023")
plt.show()
```

##**Task 2:Change in Crime Rates (2022 vs. 2023) with Unique Colors**

```python
#Using horizontal bar chart to illustrate the percentage change in
#offense rates from 2022 to 2023 for each crime type

#Offense
crime_type =["Aggravated Assault","Burglary","Larceny", "Motor Vehicle Theft",
             "Murder", "Rape", "Robbery"]
#Assign a unique color to each crime type
colors = ['deepskyblue','fuchsia','indigo','orangered','teal','seagreen','slateblue']
#Change in Offenses per 100,000 people (2022-2023)
chg_offense_rate =[-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -.8]

plt.barh(y = crime_type, width = chg_offense_rate, color = colors)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Change in Offense Rates")
plt.ylabel("Crime Type")
plt.title("Crime Rates 2022 vs. 2023")
plt.show()
```

**Why does this bar graph make more sense than Task 1's bar graph?**

Think about how absolute numbers (like total crime rates) might be misleading compared to percentage changes.
Explain why seeing the change over time is more insightful than just seeing the total number of offenses for a single year.

Visualizing the total crime rates for a single year can be misleading compared to illustrating the percentage change because the graph is only displaying data from that specific year. Without comparison to other data, it makes it difficult to identify significance of the values. Visualizing percentage change can be more useful to speculating the difference of crime rates from 2022 to 2023. In this way, you can compare how much crime rates increased or decreased over the year.

**Improve the clarity of your bar graph:**

```python
#Offense
plt.grid(axis='y') #vertical grid lines
crime_type =["Aggravated Assault","Burglary","Larceny", "Motor Vehicle Theft",
             "Murder", "Rape", "Robbery"]
colors = ['red','red','red','green','red','red','red']
#Change in Offenses per 100,000 people (2022-2023)
chg_offense_rate =[-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -.8]

plt.barh(y = crime_type, width = chg_offense_rate, color = colors)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Change in Offense Rates")
plt.ylabel("Crime Type")
plt.title("Crime Rates 2022 vs. 2023")
plt.show()
```

```python
#using vertical bar graph
plt.grid(axis='x') #horizontal grid lines
crime_type =["Aggravated Assault","Burglary","Larceny", "Motor Vehicle Theft",
             "Murder", "Rape", "Robbery"]
colors = ['red','red','red','green','red','red','red']
chg_offense_rate =[-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -.8]
plt.bar(crime_type,chg_offense_rate, color=colors)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Crime Type")
plt.ylabel("Percentage Offense Rate")
plt.title("Crime Rates 2022 vs. 2023")
plt.show()
```

##**Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023**

```python
#Combined bar chart
crime_type =["Aggravated Assault","Burglary","Larceny", "Motor Vehicle Theft",
             "Murder", "Rape", "Robbery"]
offense_rate_2023 =[264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
offense_rate_2022 =[255.4, 242.6, 1343, 330.7, -6.3, 28.2, 65.7]

plt.grid(axis='x') #horizontal grid lines
x_indices = np.arange(len(crime_type))
plt.bar(x_indices + .125,offense_rate_2023, width= .25, label='Offenses (2023)', color = 'crimson')
plt.bar(x_indices - .125,offense_rate_2022, width=.25, label='Offenses (2022)', color = 'darkorange')
plt.xticks(rotation=45, ha='right') #rotates the name of the categories
plt.xticks(ticks=x_indices,labels=crime_type)
plt.title("Crime Rates 2022 and 2023")
plt.legend()
plt.show()
```
