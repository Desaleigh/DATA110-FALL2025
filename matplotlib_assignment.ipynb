```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
# Set the default style to 'seaborn-whitegrid'
plt.style.use("seaborn-v0_8-whitegrid")
```

**Basic Line Plot**

```python
# Defining a,b, and c for the quadratic function
a = 1
b = -2
c = 1
x = np.linspace(-4,2,100)
y = a*x**2 + b*x + c

fig = plt.figure()
plt.plot(x,y)
plt.xlim([-4,2])
plt.show()
```

**Multiple Lines**

```python
x = np.linspace(0.1,2,100) #for x>0 because log(0) is undefined

y = x**3
g = np.exp(x)
h = np.log(x)

fig = plt.figure()
plt.plot(x,y, color='blue')
plt.plot(x,g, color='orange')
plt.plot(x,h, color='red')
plt.legend(['x\u00B3','e^x','log(x)'])
plt.show()
```

**Styling Lines**

```python
x = np.linspace(-1,1,100)
y = x**2
g = x**3
h = x**4
i = x**5

plt.plot(x,y, color='green', linestyle='-', linewidth=2, label='x^2')
plt.plot(x,g, color='orange', linestyle='--', linewidth=2, label='x^3')
plt.plot(x,h, color='red', linestyle='-.', linewidth=2, label='x^4')
plt.plot(x,i, color='purple', linestyle=':', linewidth=2, label='x^5')
plt.legend()
plt.show()
```

**Axes Limits and Scaling**

```python
x = np.linspace(-10,10,100) #over the range [-10,10]
y = 1/(x**2 + 1)

fig = plt.figure()
plt.plot(x,y)
plt.xlim(-10,10)
plt.ylim(0, 1)
# Demonstrate the use of plt.axis('equal')
plt.axis('equal');
plt.show()
```

```python
x = np.linspace(-10,10,100) #over the range [-10,10]
y = 1/(x**2 + 1)

# Reversed x and y axes
fig = plt.figure()
plt.plot(y,x)
plt.xlim(0,1)
plt.ylim(-10, 10)
# Demonstrate the use of plt.axis('tight')
plt.axis('tight');
plt.show()
```

**Labels and Legends**

```python
x = np.linspace(0,10,100)
y = np.sqrt(x)
g = x**3 #for x>= 0

fig = plt.figure()
plt.plot(x,y, color='purple', linestyle='--', label = '\U0000221Ax')
plt.plot(x,g, color='pink', linestyle='-', label = 'x\u00B3')
plt.title("Graph of " "\U0000221Ax" " &" " x\u00B3")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Basic Scatter Plot**

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y, 'o', color='orange')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y, color='red')
plt.show()
```

**Customizing Scatter Plots**

```python
x = np.linspace(0,4 * np.pi,100) # Range [0,4pi]
y = np.sin(x)
plt.plot(x,y, '-o', color= 'blue', markeredgecolor='purple', markeredgewidth=2, linewidth=.5, markersize=4)
plt.show()
```

**Combined Plot**

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = x*np.cos(x)

# Add scatter points fo x values at every pi/4 interval
x_point = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
y_point = x_point * np.cos(x_point)

plt.plot(x,y, '--', color='green', linewidth=2)
plt.scatter(x_point,y_point)
plt.show()
```

**Custom Plot with Alternatice Style**

```python
plt.style.use("ggplot")
```

```python
# Defining f(x) and g(x)
x = np.linspace(-np.pi/4, np.pi/4, 100)
y = np.tan(x)
g = x * np.exp(x)

plt.plot(x,y, color='green', linestyle='-', linewidth=2, label='tan(x)')
plt.plot(x,g, color='orange', linestyle='--', linewidth=2, label='xe^x')
plt.scatter(0,0)
plt.xlim(-np.pi/4, np.pi/4)
plt.ylim(-np.pi/4, np.pi/4)
plt.legend()
plt.show()
# The gray background allows the colors of the graph to pop out
# making it the main focus. In addition, including different line
# styles allows the audience to easily differentiate the functions.
```

**Style Comparison**

```python
# First Style
plt.style.use("seaborn-v0_8-white")
# Define x and y
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = x*np.cos(x)

# Add scatter points fo x values at every pi/4 interval
x_point = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
y_point = x_point * np.cos(x_point)

plt.plot(x,y, '--',linewidth=2)
plt.scatter(x_point,y_point)
plt.show()
```

```python
# Second Style
plt.style.use("seaborn-v0_8-whitegrid")

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = x*np.cos(x)

# Add scatter points fo x values at every pi/4 interval
x_point = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
y_point = x_point * np.cos(x_point)

plt.plot(x,y, '--',linewidth=2)
plt.scatter(x_point,y_point)
plt.show()
```

```python
# Third Style
plt.style.use("dark_background")

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = x*np.cos(x)

# Add scatter points fo x values at every pi/4 interval
x_point = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
y_point = x_point * np.cos(x_point)

plt.plot(x,y, '--',linewidth=2)
plt.scatter(x_point,y_point)
plt.show()
```

The first style did not have grid lines compared to the other plot styles. This design makes it difficult for the audience to pin point coordinates. On the other hand, the second style 'seaborn-whitegrid' includes gray grid lines which allows the eyes to locate the points of the graph much faster. This style keeps it minimal and simple, while still highlighting the graph of the function. Lastly, the third style, 'dark_background' creates a black background with white grid lines. This style improves the audience's readability, allowing the eyes to separate the grid from the lines/points, and makes the graph stand out more. Personally, I would use the dark background style because I like how the color pops out more with a dark bakcground compared to a light theme.
